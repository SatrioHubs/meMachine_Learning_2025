1. Manipulasi Identitas dan Deepfake (Penipuan & Misinformasi)
Tindakan ini melibatkan penggunaan AI generatif untuk meniru suara atau wajah seseorang tanpa izin untuk tujuan menipu, memfitnah, atau mempengaruhi opini publik.

Tindakan Melanggar: Menggunakan voice cloning (kloning suara) tokoh publik untuk menyebarkan pesan palsu (hoaks) atau melakukan penipuan telepon (vishing).

Kasus Nyata: Robocall Palsu Joe Biden (2024).

Detail: Menjelang pemilihan pendahuluan di New Hampshire, AS, sebuah panggilan otomatis (robocall) yang menggunakan suara AI yang meniru Presiden Joe Biden beredar. Suara tersebut meminta warga untuk tidak memberikan suara (golput).

Aspek Hukum: Komisi Komunikasi Federal AS (FCC) kemudian menyatakan bahwa panggilan suara yang dihasilkan AI tanpa izin adalah ilegal berdasarkan Telephone Consumer Protection Act.

Aspek Etika: Manipulasi demokrasi, penyebaran disinformasi, dan pencurian identitas biometrik.

Referensi:

Swenson, A., & Seitz, A. (2024). AI voice of Biden urges voters not to vote in New Hampshire primary. AP News.

FCC (Federal Communications Commission). (2024). FCC Makes AI-Generated Voices in Robocalls Illegal.

Emisi Karbon Tinggi : Proses pelatihan satu model AI besar bisa menghasilkan emisi CO2 setara dengan 5 kali lipat emisi satu mobil seumur hidupnya.

Konsumsi Air Masif : Server AI butuh pendinginan intensif. Diperkirakan percakapan pendek 20-50 pertanyaan dengan AI menghabiskan 500ml air.

Limbah Elektronik : Persaingan teknologi memicu pergantian hardware yang terlalu cepat, meningkatkan tumpukan sampah elektronik beracun.

===============================================

Solusi Mengatasinya.

Transisi ke "Green AI": Fokus mengembangkan model yang lebih kecil dan efisien (Small Language Models), bukan hanya mengejar model raksasa yang boros.

Lokasi Data Center Pintar: Membangun server di wilayah dingin (pendinginan alami) atau wilayah yang kaya energi terbarukan (air/angin) untuk menekan emisi karbon.

Hardware Khusus: Menggunakan chip yang didesain khusus AI (seperti TPU/LPU) yang jauh lebih hemat daya per komputasi dibandingkan GPU standar.
