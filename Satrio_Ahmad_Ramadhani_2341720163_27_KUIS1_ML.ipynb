# Bagian 1 - Data Loading dan Data Imputation

## Soal 1 (5 poin)
1.   Lakukan inspeksi profile data
2.   **Variabel apa** yang memiliki **nilai yang hilang** (missing value) dan **berapa** jumlahnya?



```python
df = pd.concat([X, y], axis=1)
missing_values = df.isnull().sum()
print(missing_values)
```

## Soal 2 (5 poin)
1. Lakukan proses data imputation pada fitur yang memiliki data yang hilang
2. Cek kembali apakah masih terdapat data yang hilang

```python
kolom_input = ['workclass', 'occupation', 'native-country']

for kolom in kolom_input:
    df[kolom].fillna(df[kolom].mode()[0], inplace=True)

missing_values = df.isnull().sum()
print(missing_values)
```

## Soal 3 (10 poin)
Inspeksi semua fitur kualitatif. Jika terdapat value yang **tidak sesuai**, **ganti dengan 'Others'** atau yang sesuai atau jika terdapat duplikasi karena **kesalahan penulisan**, lakukan penyesuaian.

```python
df.info()
```

```python
kolom_kualitatif = df.select_dtypes(include=['object']).columns

for kolom in kolom_kualitatif:
   df[kolom] = df[kolom].astype(str).str.strip()
   if '?' in df[kolom].unique():
       df[kolom].replace('?', 'Others', inplace=True)
print(df[kolom_kualitatif].head())
```

# Bagian 3 - Encoding Variabel Kategorical

## Soal 1 (5 poin)
Lakukan encoding pada 'Sex' dan 'Income'. 'Income' merupakan variabel target

```python
df['Sex_encode'] = df['sex'].map({'Male': 0, 'Female': 1})
df['Income_encode'] = df['income'].map({'<=50K': 0, '>50K': 1})
print(df[['sex', 'Sex_encode', 'income', 'Income_encode']])
```
